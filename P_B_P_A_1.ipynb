## Q3

```python
 s1 = float(input('Enter first side of the triangle: '))
 s2 = float(input('Enter second side of the triangle: '))
 s3 = float(input('Enter third of the triangle: '))

s = (s1 + s2 + s3) / 2 #calculating semi-perimeter

area = (s*(s-s1)*(s-s2)*(s-s3)) ** 0.5 #calculating area

print('The area of the triangle is %0.4f' %area)
```

```python
import random
l = int(input('Enter the lower bound value for the random number interval: '))
u = int(input('Enter the upper bound value for the random number interval: '))
rno=random.randint(l,u)

print("The random number generated from the interval ({0},{1}) is {2}".format(l,u,rno))
```
