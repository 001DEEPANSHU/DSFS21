## Q1

```python
#Program to find the sum of an array
n = int(input("Enter number of terms: "))
a=[]
sum=0
for i in range(1,n+1):
    a.append(float(input("Enter term number {}: ".format(i))))
print("\nEntered array is {}\n".format(a))


for i in range(0,len(a)):
    sum = sum + a[i]
print("Sum of terms of the array is {}".format(sum))
```

## Q2

```python
#Program to find the largest element in an array
n = int(input("Enter number of terms: "))
a=[]
sum=0
for i in range(1,n+1):
    a.append(float(input("Enter term number {}: ".format(i))))
print("\nEntered array is {}\n".format(a))

m = a[0]

for i in range(1,len(a)):
    if(a[i]>m):
        m=a[i]
print("The largest element in the array is {}".format(m))
```

```python
#Program for left array rotation
n = int(input("Enter number of terms: "))
a=[]
sum=0
for i in range(1,n+1):
    a.append(float(input("Enter term number {}: ".format(i))))
print("\nEntered array is {}\n".format(a))
r= int(input("Enter number of rotations: "))
r=r%len(a)
for i in range(r):
    x=a.pop(-1)
    a.insert(0,x)
print("Array after left rotation is {} ".format(a))
```

```python
#Python Program to Split the array and add the first part to the end

n = int(input("Enter number of terms: "))
a=[]
sum=0
for i in range(1,n+1):
    a.append(float(input("Enter term number {}: ".format(i))))
print("\nEntered array is {}\n".format(a))
ind= int(input("Enter index number from where you want to split the array: "))

b=[]
c=[]
for i in range(0,ind):
    b.append(a[i])
for j in range(ind,len(a)):
    c.append(a[j])

newlist = c+b
print("The resulting array is :", newlist)

```
