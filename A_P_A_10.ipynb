```python
# Write three functions:
# 1. boolean_and
# 2. boolean_or
# 3. boolean_xor
# These functions should evaluate a list of True and False values, starting from
# the leftmost element and evaluating pairwise.

def boolean_and(bool_list):
    
    for i in range(0,len(bool_list)-1):
        bool_list.insert(0,bool_list[0] and bool_list[1])
        bool_list.remove(bool_list[1])
        bool_list.remove(bool_list[1])
    
    return bool_list[0]

def boolean_or(bool_list):
    
    for i in range(0,len(bool_list)-1):
        bool_list.insert(0,bool_list[0] or bool_list[1])
        bool_list.remove(bool_list[1])
        bool_list.remove(bool_list[1])
    
    return bool_list[0]
    
    
    
def boolean_xor(bool_list):
    
    for i in range(0,len(bool_list)-1):
        bool_list.insert(0,bool_list[0] != bool_list[1])
        bool_list.remove(bool_list[1])
        bool_list.remove(bool_list[1])
    
    return bool_list[0]
    
```

```python
boolean_and([True, True, False, True])
```

```python
boolean_or([True, True, False, False]) 
```

```python
boolean_xor([True, True, False, False])
```

## Q4

```python
# Given a common phrase, return False if any individual word in the phrase 
# contains duplicate letters. Return True otherwise. 
 
def no_duplicate_letters(string):
    
    flag = True
    
    str_list = string.split(" ")
    
    for i in range(0,len(str_list)):
        if(len(str_list[i])!=len(set(str_list[i]))):
            flag =  False
            break;
            
    return flag
    
    
```

```python
no_duplicate_letters("Fortune favours the bold.")
```

```python
no_duplicate_letters("Look before you leap.") 
```

```python
no_duplicate_letters("An apple a day keeps the doctor away.") 
```
