## Q1

```python
# Program to find words which are greater than given length k
n= int(input("How many words you want to enter? "))
l=[]
for i in range(1,n+1):
    l.append(str(input("Enter word {}: ".format(i))))

l
k= int(input("\nEnter length of word: "))

print("\nWords that are longer than the given length are: ")
for i in range(0,n):
    if(len(l[i])>k):
        print(l[i])
    
```

## Q2

```python
# Program to remove the i-th character from a string

string = str(input("Enter the string: "))
k= int(input("\n Enter position of element that you want to remove:")) 

res=""
for i in range(0,len(string)):
    if(i!=k-1):
        res=res + string[i]

print("Resulting string is: ",res)
    
```

## Q3

```python
# Program to split and join a string
string = str(input("Enter the string: "))
char=input("Enter characted with which you want to join the string: ")

print("\n The splitted string is: ")
for i in range(0,len(string.split())):
    print(string.split()[i])

print("\nThe joined string is: ")
print(char.join(string.split()))
```

## Q5

```python
# Program to find uncommon words from two Strings
string1 = str(input("Enter the first string: ")).lower()
string2 = str(input("Enter the second string: ")).lower()

print("The uncommon words in the entered strings are: ")
print(list(set(string2.split()) - set(string1.split())) + list(set(string1.split()) - set(string2.split())))
```

## Q6

```python
# Program to find all duplicate characters in string
string=str(input("Enter the string: ")).lower()
strl=list(string)
elis=[]
res=[]
print("\nThe duplicate characters in the string are: \n ")
for i in range(1,len(strl)):
    elis.append(strl[i-1])
    if(strl[i] in elis):
        res.append(strl[i])

print(set(res))
```
