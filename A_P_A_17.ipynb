## Q1

```python
# Create a function that transposes a 2D matrix.

def transpose_matrix(mat):
    r = len(mat)
    c = len(mat[0])
    tmat = [[mat[i][j] for i in range(0,r)] for j in range(0,c)]
    
    for i in range(0,len(tmat)):
        print(tmat[i])
```

```python
transpose_matrix([
  [1, 1, 1],
  [2, 2, 2],
  [3, 3, 3],
  [4, 4, 4]
]) 
```

```python
transpose_matrix([
  [5, 5],
  [6, 7],
  [9, 1]
]) 
```

## Q2

```python
# Create a function that determines whether a string is a valid hex code.

def is_valid_hex_code(code):
    
    if(code[0]!="#" or len(code)!=7):
        return False
    
    else:
        
        for i in range(1,len(code)):
            if(code[i].isalnum()==False):
                return False
                break
            
        flag = True
        alpha = [i for i in code if i.isalpha() == True]
        
        for i in range(0,len(alpha)):
            if(alpha[i].lower() not in list("abcdef")):
                flag =  False
                break
        return flag
```

```python
is_valid_hex_code("#eaecee")
```

```python
is_valid_hex_code("#CD5C58C") 
```

```python
is_valid_hex_code("CD5C5C") 
```

```python
is_valid_hex_code("#CD5C&C") 
```

## Q3

```python
# Given a list of math equations (given as strings), return the percentage of correct answers 
# as a string. Round to the nearest whole number.

def mark_maths(exlist):
    import math
    
    total = len(exlist)
    count = 0
    
    for i in range(0,total):
        eqlist = exlist[i].split("=")
        if(eval(eqlist[0]) == int(eqlist[-1])):
            count = count + 1
    
    percent = str(math.ceil((count/total)*100)) + "%"
    return percent
```

```python
mark_maths(["2+2=4", "3+2=5", "10-3=3", "5+5=10"]) 
```

```python
mark_maths(["1-2=-2"])
```

```python
mark_maths(["2+3=5", "4+4=9", "3-1=2"]) 
```

## Q4

```python
def magic_square_game(l1,l2):
    alice_row = l1[0]
    alice_choice = l1[1]
    bob_col = l2[0]
    bob_choice = l2[1]
    
    sum_alice = sum([int(i) for i in alice_choice])
    sum_bob = sum([int(i) for i in bob_choice])
    
    if(sum_alice%2!=0 and sum_bob%2==0):
        if(alice_choice[bob_col-1] == bob_choice[alice_row-1]):
            return True
        else:
            return False
    else:
        return False
```

```python
magic_square_game([2, "100"], [1, "101"]) 
```

```python
magic_square_game([2, "001"], [1, "101"]) 
```

```python
magic_square_game([3, "111"], [2, "011"])
```

```python
magic_square_game([1, "010"], [3, "101"])
```

## Q5

```python
# From point A, an object is moving towards point B at constant velocity va (in km/hr). 
# From point B, another object is moving towards point A at constant velocity vb (in km/hr). 
# Knowing this and the distance between point A and B (in km), 
# write a function that returns how much time passes until both objects meet.

def lets_meet(d,va,vb):
    
    rv = va + vb
    t = d/rv
    
    sec = t*3600
    h = int(sec//3600)
    min = int((sec%3600)//60)
    sec = int((sec%3600)%60)
    
    return ("{}h {}m {}s".format(h,min,sec))
```

```python
lets_meet(280,70,80)
```

```python
lets_meet(90, 75, 65) 
```
