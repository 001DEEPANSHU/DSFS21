## Q1

```python
n = int(input("Enter the number: "))
fact = 1
for i in range(1,n+1):
    fact=fact*i
    i=i+1
print("Factorial of {} is {}".format(n,fact))
    
```

## Q4

```python
n = int(input("Enter the number: "))
lis= list(str(n))
sum = 0

for i in range(0,len(lis)):
    sum = sum + (int(lis[i]))**len(lis)
    i=i+1
if(n==sum):
    print("{} is an armstrong number".format(n))
else:
    print("{} is not an armstrong number".format(n))
```

## Q5

```python
l = int(input("Enter the lower bound of the interval: "))
u = int(input("Enter the upper bound of the interval: "))

for j in range(l,u+1):
    sum = 0
    lis= list(str(j))

    for i in range(0,len(lis)):
        sum = sum + (int(lis[i]))**len(lis)
        i=i+1
    if(j==sum):
        print("{} is an armstrong number".format(j))
    j = j+1
```

```python
n=int(input("Enter number of natural numbers to sum: "))
nt=n

if(n<0):
    print("Please enter a positive number")
else:
    sum=0
    while(n>0):
        sum = sum + n
        n = n - 1
    print("The sum of first {} natural numbers is: {}".format(nt,sum))
```
