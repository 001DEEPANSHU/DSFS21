## Q1

```python
# Create a function that takes three integer arguments (a, b, c) and returns the amount of
# integers which are of equal value.

def equal(a,b,c):
    if(len(set([a,b,c]))==3):
        return 0
    elif(len(set([a,b,c]))==1):
        return 3
    elif(len(set([a,b,c]))==2):
        return 2
```

```python
equal(3,4,3)
```

```python
equal(1,1,1)
```

```python
equal(3,4,1)
```

## Q2

```python
# Write a function that converts a dictionary into a list of keys-values tuples.

def dict_to_list(D):
    res = [i for i in D.items()]
    res.sort()
    return res

dict_to_list({"D":1,
              "B":2,
              "C":3
             })
        
```

## Q3

```python
# Write a function that creates a dictionary with each (key, value) pair being the (lower case,
# upper case) versions of a letter, respectively.

def mapping(l):
    res = {i:i.upper() for i in l}
    return res
```

```python
mapping(["a","b"])
```

```python
mapping(["a","v","y","z"])
```

## Q4

```python
# Write a function, that replaces all vowels in a string with a specified vowel.

def vow_replace(string,vow):
    for i in range(0,len(string)):
        if(string[i].lower() in ["a","e","i","o","u"]):
            string = string.replace(string[i],vow)
    
    return string

vow_replace("aaples and bananas","u")
```

## Q5

```python
# Create a function that takes a string as input and capitalizes a letter if its ASCII code is even
# and returns its lower case version if its ASCII code is odd.

def ascii_capitalize(string):
    for i in range(0,len(string)):
        if(ord(string[i])%2==0):
            string=string.replace(string[i],string[i].upper())
        else:
            string=string.replace(string[i],string[i].lower())
    return string
            
ascii_capitalize("to be or not to be")
```

```python
ascii_capitalize("THE LITTLE MERMAID")
```
