## Q1

```python
x = int(input("Enter the first number: "))  
y = int(input("Enter the second number: "))
if(x>y):
    greater = x
else:
    greater = y 
    
while(True):
    if(greater%x == 0 and greater%y == 0):
        print("LCM of {} and {} is {}".format(x,y,greater))
        break
    greater = greater + 1
```

## Q2

```python
x = int(input("Enter the first number: "))  
y = int(input("Enter the second number: "))
if(x>y):
    smaller = y
else:
    smaller = x
    
for i in range(1,smaller+1):
    if(x%i==0 and y%i==0):
        hcf = i 
    i = i + 1

print("HCF of {} and {} is {}".format(x,y,hcf))     
```

## Q3

```python
n = int(input("Enter the number:"))
print("The binary value of {} is {}".format(n,bin(n)))
print("The octal value of {} is {}".format(n,oct(n)))
print("The hexadecimal value of {} is {}".format(n,hex(n)))
```

## Q5

```python
def calculator():
    choice = int(input("""What operation do you want to perform ?
    Enter 1 for addition
    Enter 2 for substraction
    Enter 3 for multiplication
    Enter 4 for division 
    
    """))
    x = int(input("Enter the first number"))
    y = int(input("Enter the second number"))
    
    if(choice == 1):
        print("{} + {} = {}".format(x,y,x+y))
    elif(choice== 2):
        print("{} - {} = {}".format(x,y,x-y))
    elif(choice== 3):
        print("{} X {} = {}".format(x,y,x*y))
    else:
        print("{}/{} = {}".format(x,y,x/y))
```

```python
calculator()
```
