## Q1

```python
# Write a function that returns True if a given name can generate an array of words.

def anagram(word,arr):
    
    from collections import Counter
    word = word.lower()
    word_l = list(word)
    word_l.remove(" ")
    arr_word = "".join(arr)
    arr_word = arr_word.lower()
    
    if(Counter(word_l) == Counter(arr_word)):
        return True
    else:
        return False
```

```python
anagram("Justin Bieber", ["injures", "ebb", "it"])
```

```python
anagram("Chris Pratt", ["chirps", "rat"])
```

```python
anagram("Jeff Goldblum", ["jog", "meld", "bluffs"])
```

## Q2

```python
# Given an array of users, each defined by an object with the following properties: 
# name, score, reputation create a function that sorts the array to form the correct leaderboard.

# The leaderboard takes into consideration the score of each user of course, 
# but an emphasis is put on their reputation in the community, so to get the trueScore, 
# you should add the reputation multiplied by 2 to the score.
# Once you know the trueScore of each user, sort the array according to it in descending order


def leaderboards(dict_list):
    
    for i in range(0,len(dict_list)):
        dict_list[i]["trueScore"] = dict_list[i]["reputation"]*2 + dict_list[i]["score"]
    
    ts = [dict_list[i]["trueScore"] for i in range(0,len(dict_list))]
    
    tsf = ts.copy()
    tsf.sort(reverse = True)
    
    dict_list = [dict_list[ts.index(tsf[i])] for i in range(0,len(ts))]
    
    for i in range(0,len(dict_list)):
        dict_list[i].popitem()
        
    
    return dict_list
```

```python
leaderboards([
  { "name": "a", "score": 100, "reputation": 20 },
  { "name": "b", "score": 90, "reputation": 40 },
  { "name": "c", "score": 115, "reputation": 30 },
])
```

## Q3

```python
# Create a function that, given a phrase and a number of letters guessed, 
# returns a string with hyphens - for every letter of the phrase not guessed, 
# and each letter guessed in place.

def hangman(word,guessed):
    
    res = ""
    guessed = "".join(guessed)
    guessed = guessed.lower()
    
    word1 = word
    word1 = word1.lower()
    for i in range(0,len(word)):
        if(word1[i] in guessed):
            res = res + word[i]
        else:
            res = res + "-"
    
    return res
```

```python
hangman("helicopter", ["o", "e", "s"]) 
```

```python
hangman("Python rules", ["a", "n", "p", "r", "z"])
```

```python
hangman("tree", ["r", "t", "e"]) 
```

```python
# Write a function that takes in an integer n and 
# returns the highest integer in the corresponding Collatz sequence

def max_collatz(num):
    
    res = []
    res.append(num)
    
    for i in range(0,num):
        if(num%2==0):
            res.append(int(num/2))
            num = res[-1]
            if(num==1):
                break
        else:
            res.append(int(3*num + 1))
            num = res[-1]
            if(num==1):
                break
        
        
    return max(res)
```

```python
max_collatz(10)
```

```python
max_collatz(32)
```

```python
max_collatz(85)
```

```python
# Write a function that sorts a list of integers by their digit length in descending order, 
# then settles ties by sorting numbers with the same digit length in ascending order

def digit_sort(l):
    
    maxl = len(str(max(l)))
    minl = len(str(min(l)))
    
    lp = [ [] for i in range(minl,maxl+1)]
    for i in range(1,len(l)+1):
        for j in range(minl,maxl+1):
            if(len(str(l[i-1]))==j):
                lp[j-1].append(l[i-1])
    
    lp.reverse()
    for i in range(0,len(lp)):
        lp[i].sort()
    
    res = []
    
    for i in range(0,len(lp)):
        
        res = res + lp[i]
    
    return res
    
    
    
    
```
