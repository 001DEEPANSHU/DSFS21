## Q1

```python
# Given a sentence as txt, return True if any two adjacent words have this
# property: One word ends with a vowel, while the word immediately after
# begins with a vowel (a e i o u).

def vowel_links(string):
    
    flag = False
    string = string.lower()
    str_list = string.split(" ")
    
    for i in range(1,len(str_list)):
        if((str_list[i-1][-1] in ["a","e","i","o","u"]) & (str_list[i][0] in ["a","e","i","o","u"])):
            flag = True
    
    return flag
    
```

```python
vowel_links("a very large appliance") 
```

```python
vowel_links("go to edabit") 
```

```python
vowel_links("a sudden applause") 
```

```python
# Write a function that returns True if every instance of the first letter 
# occurs before every instance of the second letter

import math
def first_before_second(string,a1,a2):
    
    l1 = []
    l2 = []
    
    for i in range(0,len(string)):
        if(string[i]==a1):
            l1.append(i)

        elif(string[i]==a2):
            l2.append(i)
            
    if(max(l1)>min(l2)):
        return False
    else:
        return True
    
      
```

```python
first_before_second("precarious kangaroos", "k", "a") 
```

## Q3

```python
# Create a function that returns the characters from a list or string r on odd or even positions, 
# depending on the specifier s. The specifier will be 
# "odd" for items on odd positions (1, 3, 5, ...) and "even" for items on even positions (2, 4, 6, ...).

def char_at_pos(items,s):
    res = []
    items_list = list(items)
    
    if(s == "even"):
        for i in range(1,len(items_list),2):
            res.append(items_list[i])
        return res
    else:
        for i in range(0,len(items_list),2):
            res.append(items_list[i])
        return res
```

```python
char_at_pos([2, 4, 6, 8, 10], "even")
```

```python
char_at_pos("EDABIT", "odd")
```

```python
char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd")
```

```python
## Write a function that returns the greatest common divisor of all list elements. 
## If the greatest common divisor is 1, return 1

def GCD(items_list):

    def gcd_2_num(n1, n2):
     
        while(n2):
            n1, n2 = n2, n1 % n2
     
        return n1

    num1 = items_list[0]
    num2 = items_list[1]
    gcd = gcd_2_num(num1, num2)

    for i in range(2, len(items_list)):
        gcd = gcd_2_num(gcd, items_list[i])

    return gcd
```

```python
GCD([10,20,40]) 
```

```python
# A number/string is a palindrome if the digits/characters are the same 
# when read both forward and backward. Examples include "racecar" and 12321. 
# Given a positive number n, check if n or the binary representation of n is palindromic. 

def palindrome_type(num):
    
    num_str = str(num)
    
    num_bin = bin(num).replace("0b", "")
    
    flagd = False
    flagb = False
    
    if(num_str == num_str[::-1]):
        flagd = True
    
    if(num_bin == num_bin[::-1]):
        flagb = True
        
    if((flagd == True ) & (flagb==True)):
        return "Decimal and binary."
    elif((flagd == True ) & (flagb!=True)):
        return "Decimal only."
    elif((flagd != True ) & (flagb==True)):
        return "Binary only."
    else:
        return "Neither!"
```
