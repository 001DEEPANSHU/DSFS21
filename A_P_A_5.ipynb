## Q1

```python
# Create a function that takes a number n (integer greater than zero) as an
# argument, and returns 2 if n is odd and 8 if n is even.

import numpy as np
def f(n):
    
    k = list(np.arange(1,n+1,1))
    v = np.empty((n,),dtype=int)
    v[::2] = 2
    v[1::2] = 8
    dic = {i:j for i,j in zip(k,v)}
    return dic[n]
```

```python
f(2)
```

```python
f(1)
```

```python
f(3)
```

## Q2

```python
# Create a function that returns the majority vote in a list. A majority vote is
# an element that occurs &gt; N/2 times in a list (where N is the length of the list).

def majority_vote(l):
 
    maxcount = 0
    n = len(l)
    ind = -1
    for i in range(n):
        count = 0
        for j in range(n):
            if(l[i] == l[j]):
                count += 1
        if(count > maxcount):
            maxcount = count
            ind = i
    if (maxcount > n//2):
        return l[ind]
    else:
        return "None" 
```

```python
majority_vote(["A","B","B","A","C","C"])
```

```python
majority_vote(["A","A","B"])
```

## Q3

```python
# Create a function that takes a string txt and censors any word from a given
# list lst. The text removed must be replaced by the given character char.

def censor_string(txt,lst,char):
    
    txt_l = txt.split(" ")
    
    for i in range(0,len(txt_l)):
        if(txt_l[i] in lst):
            for j in range(0,len(txt_l[i])):
                txt_l[i]= txt_l[i].replace(txt_l[i][j],char)
                                
    return " ".join(txt_l)
```

```python
censor_string("The cow jumped over the moon.", ["cow", "over"], "*")
```

```python
censor_string("Why did the chicken cross the road ?", ["did", "chicken", "road"], "*") 
```

## Q4

```python
# Create a function which takes an integer n and returns True if the given
# number is a Polydivisible Number and False otherwise.

def is_polydivisible(num):
    
    num_str = str(num)
    
    if(num_str[0]==0):
        return False
    
    else:
        c = 0 
        for i in range(1,len(num_str)+1):
            
            if(int(num_str[0:i])%i == 0):
                c = c + 1
        
        if(c == len(num_str)):
            return True
        else:
            return False    
```

```python
is_polydivisible(1232)
```

```python
is_polydivisible(123220)
```

```python
# Create a function that takes a list of numbers and returns the sum of all
# prime numbers in the list.

def sum_primes(num_list):
    
    def prime(n):
        
        flag = False
        if(n>1):

            for i in range(2,n):

                if(n%i==0):
                    flag = True
                    break

        return flag

    sum = 0
    
    for i in range(0,len(num_list)):
        
        if(prime(num_list[i])==False):
            sum = sum + num_list[i]
    
    return sum

            
                    
```
