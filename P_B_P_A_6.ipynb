## Q1

```python
# Recursive function to generate a fibonacci sequence
def fibo_series(n):
    if(n<=1):
        return(n)
    else:
        return(fibo_series(n-1)+fibo_series(n-2))

n=int(input("Enter no. of terms:"))
if(n<=0):
    print("Enter a positive integer")
else:
    print("Fibonacci series is: ")
    fseries=[]
    for i in range(0,n):
        fseries.append(fibo_series(i))
    print(fseries)
        
```

## Q2

```python
# Recursive function to caculate a factorial
def factorial(n):
    if(n==0):
        return(1)
    else:
        return(n*factorial(n-1))

factorial(int(input("Enter the number to caculate the factorial:")))
```

```python
# Program to calculate the natural log of a number
import math
n=float(input("Enter the number: "))
if(n<=0):
    print("Enter a positive number")
else:
    print("The natural log of {0} is {1:.4f}".format(n,math.log(n)))
```

## Q5

```python
# Program to sum cubes of first n natural numbers
n=int(input("Enter the number of terms: "))
sum = 0
for i in range(1,n+1):
    sum=sum + i**3
    i=i+1

print("The sum of cubes of first {} natural numbers is {}".format(n,sum))
```
