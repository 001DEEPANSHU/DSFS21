## Q1

```python
# Create a function that takes a list and string. The function should remove
# the letters in the string from the list, and return the list.

def remove_letters(l,string):
    
    from collections import Counter
    
    counter1 = Counter(l)
    counter2 = Counter(string)
    
    res_counter = counter1 - counter2
    
    return list(res_counter)
```

```python
remove_letters(["s", "t", "r", "i", "n", "g", "w"], "string")
```

```python
remove_letters(["b", "b", "l", "l", "g", "n", "o", "a", "w"], "balloon")
```

```python
remove_letters(["d", "b", "t", "e", "a", "i"], "edabit") 
```

## Q2

```python
# Create a function that takes a number (step) as an argument and returns the amount of blocks in that step.

def blocks(steps):
    
    blocks = 5
    
    if(steps==1):
        return blocks
    
    else:
        n = 7
        for i in range(1,steps):
            blocks = blocks + n
            n = n + 1

                
    return blocks
```

```python
blocks(1)
```

```python
blocks(3)
```

```python
blocks(5)
```

## Q3

```python
# Create a function that subtracts one positive integer from another, 
# without using any arithmetic operators such as -, %, /, +, etc.

def my_sub(num1,num2):
    
    res = [i for i in range(num1,num2)]
    return len(res)
    
    
```

```python
my_sub(0,0)
```

```python
my_sub(5,9)
```

```python
my_sub(10,30)
```

## Q4

```python
# Create a function that takes a string containing money in dollars and pounds sterling
# (seperated by comma) and returns the sum of dollar bills only, as an integer.

def add_bill(string):
    
    l_str = string.split(",")
    
    s = 0
    
    for i in range(0,len(l_str)):
        if(l_str[i][0] == "d"):
            s = s + int(l_str[i][1:])
    return s
        
    
```

```python
add_bill("d20,p40,p60,d50")
```

```python
add_bill("p30,d20,p60,d150,p360") 
```

```python
# Create a function that flips a horizontal list into a vertical list, 
# and a vertical list into a horizontal list.

def flip_list(l):
    
    res = []
    if(l == []):
        return []
    
    elif(type(l[0])==list):
        for i in range(0,len(l)):
            res.append(l[i][0])
    
        return res
    
    else:
        for i in range(0,len(l)):
            res.append([l[i]])
        return res
        
    

    
```

```python
flip_list([[5], [6], [9]])
```

```python
flip_list([1, 2, 3, 4])
```

```python
flip_list([])
```
