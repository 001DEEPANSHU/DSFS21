## Q1

```python
# YouTube offers different playback speed options for users. This allows
# users to increase or decrease the speed of the video content. Given the
# actual duration and playback speed of the video, calculate the playback
# duration of the video.

def playback_duration(duration,speed):
    
    duration = duration.split(":")
    h,m,s = int(duration[0]),int(duration[1]),int(duration[2])
    
    ts = h*3600 + m*60 + s
    ns = ts//speed
    
    nh = int(ns//3600)
    nm = int((ns%3600)//60)
    ns = int(((ns%3600)%60))
    
    if(nh==0):
        nh="00"
    if(nm==0):
        nm="00"
    if(ns==0):
        ns="00"
    
    
    return ("{}:{}:{}".format(nh,nm,ns))
```

```python
playback_duration("02:30:00",2)
```

```python
playback_duration("01:20:00", 1.5)
```

```python
playback_duration("00:30:00", 2) 
```

```python
playback_duration("51:20:09", 0.5) 
```

```python
# The cube at the bottom will have a volume of n^3, the cube above will have volume of (n-1)^3 and so on 
# until the top which will have a volume of 1^3.
# Given the total volume m of the building, can you find the number of cubes n required for the building?

def pile_of_cubes(vol):
    
    if(vol==1):
        return 1
    
    else:
        n = 2
        cvol = 1
        
        while(cvol<vol):
            cvol = cvol + n**3
            
            if(cvol==vol):
                return n
            
            else:
                n = n +1
                continue
        
        return "None" 
```

```python
pile_of_cubes(107)
```

```python
pile_of_cubes(4183059834009)
```

```python
pile_of_cubes(1071225)
```

## Q3

```python
# A fulcrum of a list is an integer such that all elements to the left of it 
# and all elements to the right of it sum to the same value. Write a function that finds the fulcrum of a list.

def find_fulcrum(l):
    
    for i in range(0,len(l)):
        if(sum(l[0:i])==sum(l[i+1:])):
            return l[i]
    return -1
```

```python
find_fulcrum([1,1,2,4,-2])
```

```python
find_fulcrum([1, 2, 4, 9, 10, -10, -9, 3]) 
```

```python
find_fulcrum([8, 8, 8, 8])
```

## Q4

```python
# Given a list of integers representing the color of each sock, determine how many 
# pairs of socks with matching colors there are. For example, there are 7 socks with colors [1, 2, 1, 2, 1, 3, 2]. 
# There is one pair of color 1 and one of color 2. 
# There are three odd socks left, one of each color. The number of pairs is 2.
# Create a function that returns an integer representing the number of matching pairs 
# of socks that are available.

def sock_merchant(l):
    
    pairs = 0 
    s = list(set(l))
    for i in range(0,len(s)):
        pairs = pairs + l.count(s[i])//2
    
    return pairs
        
        
```

```python
sock_merchant([10, 20, 20, 10, 10, 30, 50, 10, 20])
```

```python
sock_merchant([50, 20, 30, 90, 30, 20, 50, 20, 90])
```

```python
sock_merchant([]) 
```

```python
# Create a function that takes a string containing integers 
# as well as other characters and return the sum of the negative integers only.

def negative_sum(char_string):
    
    import re
    regob = re.compile(r"(-\d+)")
    
    num_list = regob.findall(char_string)
    
    int_num_list = [int(i) for i in num_list]
    
    return sum(int_num_list)
    
```

```python
negative_sum("-12 13%14&-11") 
```

```python
negative_sum("22 13%14&-11-22 13 12") 
```
