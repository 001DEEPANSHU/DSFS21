## Q1

```python
# Create a function that takes a list and returns a new list containing only
# prime numbers.

def filter_primes(num_l):
    
    l = num_l.copy()
    for i in l:
       # print(i)
        if(i==2):
            continue
        else:
            for j in range(2,i):
                if(i%j==0):
                    num_l.remove(i)
                    break
    return num_l
                
```

```python
filter_primes([7, 9, 3, 9, 10, 11, 27]) 
```

```python
filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70]) 
```

```python
filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093, 1097]) 
```

## Q2

```python
# Once a water balloon pops, is soaks the area around it. 
# The ground gets drier the further away you travel from the balloon.
#The effect of a water balloon popping can be modeled using a list. 
# Create a function that takes a list which takes the pre-pop state and 
# returns the state after the balloon is popped. The pre-pop state will contain at most a single balloon, 
# whose size is represented by the only non-zero element.

def pop(l):
    
    res = []
    
    ind = int((len(l) - 1)/2)
    
    for i in range(1,ind+1):
        res.append(l[ind]-i)
    
    res.reverse()
    res.append(l[ind])
    
    for i in range(len(res)-2,-1,-1):
        res.append(res[i])
    
    return res
```

```python
pop([0, 0, 0, 0, 4, 0, 0, 0, 0])
```

```python
pop([0])
```

```python
pop([0, 0, 0, 3, 0, 0, 0])
```

## Q3

```python
# Given a number of petals, return a string which repeats the phrases "Loves me" and "Loves me not" 
# for every alternating petal, and return the last phrase in all caps. 
# Remember to put a comma and space between phrases.

def loves_me(num):
    
    res = []
    
    for i in range(1,num+1):
        if(i%2!=0):
            res.append("Loves me")
        else:
            res.append("Loves me not")

    res[-1] = res[-1].upper()
    
    
    return ", ".join(res)
```

```python
loves_me(3)
```

```python
loves_me(6) 
```

```python
loves_me(1) 
```

```python
# Write a function that sorts each string in a list by the letter in alphabetic ascending order (a-z).

def sort_by_letter(l):
    
    res = []
    
    for i in range(0,len(l)):
        for j in range(0,len(l[i])):

            if(str(l[i][j]).isalpha()==True):
                res.append(l[i][j])
    
    eres = res.copy()
    res.sort()
    return [l[eres.index(res[j])] for j in range(0,len(l))]
    
    
    
```

```python
sort_by_letter(["932c", "832u32", "2344b"])
```

```python
sort_by_letter(["572z", "5y5", "304q2"])
```

```python
sort_by_letter(["99a", "78b", "c2345", "11d"])
```

```python
# Create a function that returns the letter position that the ball is at, 
# once I finish swapping the cups. The swaps will be given to you as a list.


def cup_swapping(swap_list):
    
    position = 'B'
    
    for i in swap_list:
        
        if(position in i):
            
            if(i[0]==position): 
                position = i[1]
            else:
                position = i[0]   
    return position
    
```

```python
cup_swapping(["AB", "CA", "AB"]) 
```
