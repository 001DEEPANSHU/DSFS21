## Q1

```python
# 1. Create a function to perform basic arithmetic operations that includes
# addition, subtraction, multiplication and division on a string number

def arithmatic_operation(exp):
    exp=exp.split(" ")
    o = exp[1]
    a = int(exp[0])
    b = int(exp[2])
    
    if(o =="+"):
        return a + b 
    elif(o == "-"):
        return a - b
    elif(o == "*"):
        return a*b
    elif(o == "//" and b > 0):
        return a/b
    else: 
        return -1
    
```

```python
arithmatic_operation("20 // 0")
```

```python
arithmatic_operation("12 * 12")
```

## Q2

```python
# Write a function that takes the coordinates of three points in the form of a
# 2d array and returns the perimeter of the triangle. The given points are the
# vertices of a triangle on a two-dimensional plane.

import math

def perimeter(co_list):
    p1,p2,p3 = co_list[0],co_list[1],co_list[2]
    
    x1,y1 = p1[0],p1[1]
    x2,y2 = p2[0],p2[1]
    x3,y3 = p3[0],p3[1]
    
    A = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    B = math.sqrt((x3-x1)**2 + (y3-y1)**2)
    C = math.sqrt((x3-x2)**2 + (y3-y2)**2)
    
    P = round(A + B + C,2)
    return P
```

```python
perimeter([[15, 7], [5, 22], [11, 1]])
```

## Q3

```python
# Create a function that takes a skyline (2-D list of 0&#39;s and 1&#39;s) and returns the
# height of the tallest skyscraper.

def tallest_skyscraper(l):
    
    d = []
    for j in range(0,len(l[0])):
        sum = 0 
        for i in range(0,len(l)):
            sum = sum + l[i][j]
        d.append(sum)

    return(max(d))
```

```python
tallest_skyscraper([
[0, 1, 0, 0],
[0, 1, 0, 0],
[0, 1, 1, 0],
[1, 1, 1, 1]
])
```

```python
# Write a function to read the billable days of an employee and return the bonus
# he/she has obtained in that quarter.

def bonus(days):
    
    if(days < 33):
        b = 0
        c = 0 
        d = 0 
    elif(33 <= days <= 40):
        b = days - 32
        c = 0
        d = 0 
    elif(41 <= days <=48):
        b = 8 
        c = days - 40
        d = 0
    elif(days > 48):
        b = 8
        c = 8
        d = days - 48
    
    return(b*325 + c*550 + d*600)
        
```

```python
# Create a function that determines whether a number is a Disarium or not.

def is_disarium(num):
    stnum = str(num)
    sum = 0
    for i in range(0,len(stnum)):
        sum = sum + int(stnum[i])**(i+1)
    
    if(sum==num):
         return True
    else:
        return False
```

```python
is_disarium(544)
```

```python
is_disarium(518)
```
