```python
km = float(input("Enter distance in kilometers: "))
miles = 0.621371*km
print("{0} kilometers is {1} miles".format(km,miles))
```

```python
cel = float(input("Enter temperature in Celsius: "))
fahr = (cel * 9/5) + 32
print("{0} degree Celsius is equal to {1} degree Fahrenheit".format(cel,fahr))
```

## Q4

```python
import cmath

a = float(input("Enter the value of coefficient of x^2: "))
b = float(input("Enter the value of coefficient of x: "))
c = float(input("Enter the value of constant : "))

d = (b**2) - (4*a*c) # calculating the Discriminant

x1 = (-b - cmath.sqrt(d))/(2*a)
x2 = (-b + cmath.sqrt(d))/(2*a)

print("The solutions for the given quadratic equation are {0} and {1}".format(x1,x2))
```
