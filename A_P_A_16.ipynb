## Q1

```python
# Create a function which validates whether a given string is a valid Rondo
# Form.

def valid_rondo(string):
    
    if(string[0] != "A" or string[-1]!="A"):
        return False
    
    elif(len(string)<3):
        return False
    
    else:
        
        alpha = list("BCDEFGHIJKLMNOPQRSTUVWXYZ")
        if(list(string[1::2])== alpha[0:len(string[1::2])]):
            return True
        else:
            return False



```

```python
valid_rondo("ABACADAEAFAGAHAIAJA")
```

```python
valid_rondo("ABA") 
```

```python
valid_rondo("ABBACCA")
```

```python
valid_rondo("ACAC")
```

```python
valid_rondo("A") 
```

## Q2

```python
# Create a function that returns the whole of the first sentence which contains a specific word. 
# Include the full stop at the end of the sentence.

def sentence_searcher(text,word):
    
    
    text1 = text.lower()
    word = word.lower()
    txt_list = text.split(".")
    txt_list1 = text1.split(".")
    
    for i in range(0,len(txt_list1)):
        if(word in txt_list1[i]):
            return txt_list[i]+"."
            break
    
    
    
```

```python
txt = "I have a cat. I have a mat. Things are going swell."
```

```python
sentence_searcher(txt, "have") 
```

```python
sentence_searcher(txt, "MAT") 
```

```python
# Given a number, find the "round "of each digit of the number. 
# An integer is called "round" if all its digits except the leftmost (most significant) are equal to zero.

def sum_round(digit):
    
    
    res = []
    for i in range(0,len(str(digit))):
        l = len(str(digit))-1
        a = (digit//(10**l))*(10**l)

        digit = digit - a
        res.append(str(a))
      
    res.reverse()
    res = " ".join(res[:len(res)])


    if(res[0]=="0"):
        
        return res[2:]
    else:
        return res


```

```python
# Your task, is to create N x N multiplication table, of size n provided in parameter.

def multiplication_table(n):
    
    table = [[] for i in range(0,n)]
    
    for i in range(0,n):
        for j in range(0,n):
            table[i].append((i+1)*(j+1))
    
    return table
```

```python
multiplication_table(3)
```

```python
multiplication_table(1)
```

```python
# Create a function that returns True if two lines rhyme and False otherwise. 
# For the purposes of this exercise, two lines rhyme if the last word from each sentence contains the same vowels.

def does_rhyme(line1,line2):
    
    l1 = line1.split(" ")
    l2 = line2.split(" ")
    
    lw1 = l1[-1]
    lw2 = l2[-1]
    
    v1 = []
    for i in lw1:
        if(i.lower() in list("aeiou") ):
            v1.append(i.lower())
    
    v2 = []
    for i in lw2:
        if(i.lower() in list("aeiou") ):
            v2.append(i.lower())
    
    if(v1==v2):
        return True
    else:
        return False
```

```python
does_rhyme("Sam I Am!", "Green eggs and ham.") 
```

```python
does_rhyme("Sam I am!", "Green eggs and HAM.")
```

```python
does_rhyme("You are off to the races", "a splendid day.")
```

```python
does_rhyme("and frequently do?", "you gotta move.")
```
