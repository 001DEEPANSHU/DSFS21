```python
# Given a sentence, return the number of words which have the same first
# and last letter.

def count_same_ends(string):
    
    string_list = string.lower().split(" ")
    
    count = 0
    for i in range(0,len(string_list)):
            if(len(string_list[i])>1):
                if(string_list[i][-1].isalpha()==False):

                    if(string_list[i][0]==string_list[i][-2]):
                        count = count + 1
                else:
                    if(string_list[i][0]==string_list[i][-1]):
                        count = count + 1

    return count
            
```

```python
count_same_ends("Pop! goes the balloon")
```

```python
count_same_ends("No I am not in a gang.")
```

```python
count_same_ends("And the crowd goes wild!")
```

```python
# Create a function that takes a string and applies the Atbash cipher to it.

def atbash(s):
    
    import string
    lower_alpha = list(string.ascii_lowercase)
    upper_alpha = list(string.ascii_uppercase)
    
    s_list = list(s)
    for i in range(0,len(s)):
        if(s[i].isalpha()==True):
            if(s[i].islower()==True):
                s_list[i] = lower_alpha[25 - lower_alpha.index(s_list[i])]
            else:
                s_list[i] = upper_alpha[25 - upper_alpha.index(s_list[i])]
        else:
            continue

    return "".join(s_list)
```

```python
atbash("Hello world!")
```

```python
atbash("apple") 
```

```python
# Create a class Employee that will take a full name as argument, as well as a set of none, one or more keywords. 
# Each instance should have a name and a lastname attributes plus one more attribute for each of the keywords, 
# if any.


class Employee(object):

    def __init__(self,name,**kwargs):
        self.name = name
        self.firstname = name.split(" ")[0]
        self.lastname = name.split(" ")[1]

        for i in range(0,len(kwargs)):
            setattr(self,list(kwargs.keys())[i],kwargs[list(kwargs.keys())[i]])
        


```

```python
# Create a function that determines whether each seat can "see" the front-stage. 
# A number can "see" the front-stage if it is strictly greater than the number before it.


def can_see_stage(a):
    
    
    r = len(a)
    c = len(a[0])
    res = []
    for j in range(0,c):
        l=[]
        for i in range(0,r):
            l.append((a[i][j]))
        res.append(l)
    
    for i in range(1,len(res)+1):
        if(res[i]<res[i-1]):
            return False
            break
        else:
            return True
```

```python
can_see_stage([[1, 2, 3, 2, 1, 1],
[2, 4, 4, 3, 2, 2],
[5, 5, 5, 5, 4, 4],
[6, 6, 7, 6, 5, 5]])
```

```python
can_see_stage([
  [1, 0, 0],
  [1, 1, 1],
  [2, 2, 2]
])
```

```python
can_see_stage([
  [2, 0, 0], 
  [1, 1, 1], 
  [2, 2, 2]
]) 
```

```python
can_see_stage([
  [0, 0, 0],
  [1, 1, 1],
  [2, 2, 2]
]) 
```

```python
# Create a Pizza class with the attributes order_number and ingredients (which is given as a list). 
# Only the ingredients will be given as input.

# You should also make it so that its possible to choose a ready made pizza flavour rather than 
# typing out the ingredients manually! As well as creating this Pizza class, 
#hard-code the following pizza flavours.

# Name	              Ingredients
# hawaiian	          ham, pineapple
# meat_festival	      beef, meatball, bacon
# garden_feast	      spinach, olives, mushroom


class Pizza:
    obj_number = 0
    ingredients = []
    
    def __init__(self,ingredients):

        self.ingredients = ingredients
        Pizza.obj_number = Pizza.obj_number + 1
        setattr(self,"order_number",Pizza.obj_number)
        

    def garden_feast():
        return Pizza([ "spinach","olives", "mushroom"])
    
    def meat_festival():
        return Pizza([ "beef","meatball", "bacon"])

    def hawaiian():
        return Pizza([ "ham","pineapple"])
```
