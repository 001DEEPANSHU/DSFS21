## Q1

```python
#Program to add 2 matrices
r = int(input("Enter number of rows: "))
c = int(input("Enter number of columns: "))
m1 = []
m2 = []

print("\n Enter elements for the first matrix: ")
for i in range(0,r):
    row=[]
    for j in range(0,c):
        row.append(int(input("Enter element {} for row {}:  ".format(j+1,i+1))))
    m1.append(row)

print("\n Enter elements for the second matrix: ")
for i in range(0,r):
    row=[]
    for j in range(0,c):
        row.append(int(input("Enter element {} for row {}:  ".format(j+1,i+1))))
    m2.append(row)

summat=[]

for i in range(r):
    row = []
    for j in range(c):
        row.append(m1[i][j] + m2[i][j])
    summat.append(row)

print("\n The sum of the matrices is: ")
for i in range(r):
    print(summat[i])
```

## Q2

```python
#Program to Multiply Two Matrices

r1 = int(input("Enter number of rows for the first matrix: "))
c1 = int(input("Enter number of columns for the first matrix: "))
m1 = []
m2 = []

print("\n Enter elements for the first matrix: ")
for i in range(0,r1):
    row=[]
    for j in range(0,c1):
        row.append(int(input("Enter element {} for row {}:  ".format(j+1,i+1))))
    m1.append(row)

r2 = int(input("Enter number of rows for the second matrix: "))
c2 = int(input("Enter number of columns for the second matrix: "))

print("\nEnter elements for the second matrix: ")
for i in range(0,r2):
    row=[]
    for j in range(0,c2):
        row.append(int(input("Enter element {} for row {}:  ".format(j+1,i+1))))
    m2.append(row)

product = []
if(c1!=r2):
    print("\nInvalid matrices dimensions for matrix multiplication ")
else:
    for i in range(0,r1):
        row=[]
        for j in range(0,c2):
            a=0
            for k in range(0,r2):
                a=a+m1[i][k]*m2[k][j]
            row.append(a)
        product.append(row)

print("\nThe product of the given matrices is: \n")
for i in range(r1):
    print(product[i])    
```

## Q3

```python
# Program to transpose a matrix
r = int(input("Enter number of rows: "))
c = int(input("Enter number of columns: "))

m=[]

print("\n Enter elements for the first matrix: ")
for i in range(0,r):
    row=[]
    for j in range(0,c):
        row.append(int(input("Enter element {} for row {}:  ".format(j+1,i+1))))
    m.append(row)

tm = [] 
for j in range(0,c):
    row=[]
    for i in range(0,r):
        row.append(m[i][j])
    tm.append(row)

print("\n The transpose of the given matrix is: ")
for i in range(0,c):
    print(tm[i])
```

## Q4

```python
# Program to sort words in alphabatical order
```

```python
n = int(input("Enter the number of words: "))
l=[]
for i in range(0,n):
    l.append(str(input("Enter word number {}: ".format(i+1))))

print("\n Entered words are: ",l)

for i in range(0,len(l)):
    l[i] = l[i].lower()

l.sort()
print("\n Sorted words are: ",l)
```

## Q5

```python
# Program to Remove Punctuation From a String

punct = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
sentence = str(input("Enter the sentence: "))

result = ""

for i in sentence:
    if(i not in punct):
        result = result+i

print("\n ",result)
```
