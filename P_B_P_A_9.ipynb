## Q1

```python
# Program to check if the given number is a Disarium Number?
n=int(input("Enter the number: "))
sum = 0
for i in str(n):
    sum = sum + int(i)**(list(str(n)).index(i)+1)

if(n==sum):
    print("\n{} is a Disarium number".format(n))
else: 
    print("\n {} is not a Disarium number".format(n))
```

## Q3

```python
# Program to check if a given number is a happy number

n=int(input("Enter the number: "))


def check_happy(n):
    sum = 0
    
    for i in str(n):
        sum = sum + int(i)**2
    if(sum==1 or sum==7):
        print("a happy number")

    elif(sum in (2,3,4,5,6,8,9)): 
        print("not a happy number")

    else:
        check_happy(sum)


print("{} is ".format(n),end="") 
check_happy(n)
```

## Q5

```python
# Program to determine whether the given number is a Harshad Number


def check_harshad(n):
    sum = 0 
    num=n 

    while(num>0):
        r=num%10 
        sum=sum+r
        num=num//10  

    if(n%sum==0):
        print("{} is a Hashad number".format(n))
    else: 
        print("{} is not a Hashad number".format(n))

check_harshad(int(input("Enter the number: ")))
```

```python
# Program to print all pronic numbers between 1 and 100

def check_pronic(num):

    for i in range(1,num):
        if(i*(i+1)==num):
            print("{} is a pronic number ".format(num))

for j in range(1,101):
    check_pronic(j)
```
