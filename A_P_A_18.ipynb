## Q1


```python
# A robot has been given a list of movement instructions. Each instruction is either left, right, up or down, 
# followed by a distance to move. The robot starts at [0, 0]. You want to calculate where the robot will end up 
# and return its final position as a list

def track_robot(ins_list):
    x = 0
    y = 0
    
    for i in range(0,len(ins_list)):
        ins = ins_list[i].split(" ")
        
        if(ins[0]=="right"):
            x = x + int(ins[-1])
        elif(ins[0]=="left"):
            x = x - int(ins[-1])
        elif(ins[0]=="up"):
            y = y + int(ins[-1])
        else:
            y = y - int(ins[-1])
    
    return [x,y]
```

```python
track_robot(["right 10", "up 50", "left 30", "down 10"])
```

```python
track_robot([])
```

```python
track_robot(["right 100", "right 100", "up 500", "up 10000"]) 
```

```python
# Write a function that will return the longest word in a sentence. 
# In cases where more than one word is found, return the first one.

def find_longest(sent):
    
    sent_list = sent.split(" ")
    
    temp = []
    for i in range(0,len(sent_list)):
        a = ""
        for j in range(0,len(sent_list[i])):
            if(sent_list[i][j].isalpha()==True):
                a = a + sent_list[i][j]
        temp.append(a)
    
    len_list = [len(i) for i in temp]
    
    for i in range(0,len(temp)):
        if(len(temp[i])==max(len_list)):
            return temp[i].lower()
            break
```

```python
find_longest("Forgetfulness is by all means powerless!")
```

```python
find_longest("A thing of beauty is a joy forever.") 
```

```python
find_longest("\"Strengths\" is the longest and most commonly used word that contains only a single vowel.") 
```

```python
# Write a function that divides a list into chunks of size n, where n is the length of each chunk.

def chunkify(l,chunks):
    import math
        
    flen = math.ceil(len(l)/chunks)
    
    res = []
    
    if(len(l)%2==0):
        for j in range(0,flen):
            temp = []
            for  i in range(0,chunks):
                temp.append(l[i])

            l = l[chunks:]
            res.append(temp)
            
    else:
        for j in range(0,flen-1):
            temp = []
            for  i in range(0,chunks):
                temp.append(l[i])

            l = l[chunks:]
            res.append(temp)

        res.append([l[-1]])
    return res
```

```python
chunkify([2, 3, 4, 5], 2)
```

```python
chunkify([2, 3, 4, 5, 6, 7], 3)
```

```python
chunkify([2, 3, 4, 5, 6], 2) 
```

```python
# You are given a list of strings consisting of grocery items, with prices in parentheses. 
# Return a list of prices in float format.

def get_prices(l):
    
    res = []    
    for i in range(0,len(l)):
        temp = []
        temp = l[i].split(" ")[-1].split("$")[-1].split(')')[0]
        res.append(eval(temp))
    return res
```

```python
get_prices([
  "artichokes ($1.99)",
  "rotiserrie chicken ($5.99)",
  "gum ($0.75)"
])
```
