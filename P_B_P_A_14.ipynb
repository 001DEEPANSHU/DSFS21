## Q1


```python
#Define a class with a generator which can iterate the numbers, which are divisible by
#7, between a given range 0 and n.

class iter_class(object):
    def __init__(self,n):
        self.n = n
    
    def div_func(self):
        for i in range(0,self.n):
            if(i%7==0):
                yield(i)

n = int(input("Enter the value of n:"))

print("The numbers that are divisible by 7 are: ")

for i in iter_class(n).div_func():
    print(i)
```

```python
# Program to compute the frequency of the words from the input. The output
# should output after sorting the key alphanumerically.
import pandas as pd

word_str = str(input("Enter the sentence:"))

words_list = word_str.split(" ")

words_list = pd.Series(words_list)


print("\nThe frequency distribution of the words is given below:\n")
print(words_list.value_counts().sort_index())
```

## Q3


```python
#Define a class Person and its two child classes: Male and Female. All classes have a
#method "getGender" which can print "Male" for Male class and "Female" for Female
#class.


class Person(object):
    def __init__(self,gender="Unknown"):
        self.gender = "Unknown"
    def getGender( self ):
        return self.gender

class Male( Person ):
    def __init__(self):
        self.gender = "Male"
    
    def getGender( self ):
        return self.gender

class Female( Person ):
    def __init__(self):
        self.gender = "Female"
    
    def getGender( self ):
        return self.gender

male = Male()
female = Female()
print(male.getGender())
print(female.getGender())
```

## Q4

```python
#Please write a program to generate all sentences where subject is in ["I", "You"] and
#verb is in ["Play", "Love"] and the object is in ["Hockey","Football"].


sub_list=["I", "You"]
verbs_list=["Play", "Love"]
obj_list=["Hockey","Football"]

sen_list = [(sub +" "+ vb + " " + ob) for sub in sub_list for vb in verbs_list for ob in obj_list]

for sen in sen_list:
    print(sen)
```

```python
# Writing a program to compress and decompress the string "hello world!hello
# world!hello world!hello world!"

import zlib
string = 'hello world!hello world!hello world!hello world!'
bt = zlib.compress(bytes(string,'utf-8'))
print(bt)

print("\n")

print(zlib.decompress(bt))
```

```python
# Writing a binary search function which searches an item in a sorted list. The
# function should return the index of element to be searched in the list.

n = int(input("Enter number of elements in the list:"))
l=[]

for i in range(0,n):
    l.append(input("Enter element {}:".format(i+1)))

print("The entered list is: ",l)

l.sort()

print("The sorted list is: ",l)

char=input("Which element you want to search?")

for i in range(0,n):
    if(l[i] == char):
        print("The index for element {} is {}".format(char,i))

```
