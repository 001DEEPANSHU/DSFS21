## Q1

```python
# Program to find the sum of elements in a list
n=int(input("Enter no. of elements in the list: "))

l=[]
sum = 0
for i in range(1,n+1):
    l.append(float(input("Enter element {} :".format(i))))
    sum = sum + l[i-1]

print("\nEntered list is: ", l)
print("\nSum of the elements in the list is: ", sum)
```

## Q2

```python
# Program to find the product of all the elements in a list
n=int(input("Enter no. of elements in the list: "))

l=[]
p = 1
for i in range(1,n+1):
    l.append(float(input("Enter element {} :".format(i))))
    p = p*l[i-1]

print("\nEntered list is: ", l)
print("\Product of all the elements in the list is: ", p)
```

## Q4

```python
# Program to find the largest number in a list
n=int(input("Enter no. of elements in the list: "))

l=[]

for i in range(1,n+1):
    l.append(float(input("Enter element {} :".format(i))))

max = l[0]
for i in range(0,n):
    if(l[i]>max):
        max=l[i]

print("\nEntered list is: ", l)
print("\nThe largest element in the list is: ", max)
```

## Q7

```python
# Program to find the even numbers in a list

n=int(input("Enter no. of elements in the list: "))

l=[]

for i in range(1,n+1):
    l.append(float(input("Enter element {} :".format(i))))

print("\nEntered list is: ", l)
print("\nEven numbers in the list are: ")

for i in l:
    if(i%2==0):
        print(i)
```

## Q10

```python
#Program to clone a list

n=int(input("Enter no. of elements in the list: "))

l=[]

for i in range(1,n+1):
    l.append(float(input("Enter element {} :".format(i))))

cl=[]

for i in range(0,n):
    cl.append(l[i])

print("\nThe cloned list is: ",cl)    
```

## Q11

```python
# Program to count the occurence of an element in the list

n=int(input("Enter no. of elements in the list: "))

l=[]

for i in range(1,n+1):
    l.append(float(input("Enter element {} :".format(i))))

e = float(input("\nEnter which element to search for: "))
count = 0 
for i in range(0,n):
    if(l[i]==e):
        count = count+ 1
print("\n{} occurs {} times in the list".format(e,count))
    
```
