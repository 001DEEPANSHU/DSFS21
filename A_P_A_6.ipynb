## Q2

```python
# Given a function that accepts unlimited arguments, check and count how
# many data types are in those arguments. Finally return the total in a list.
# List order is:
# [int, str, bool, list, tuple, dictionary]

def count_datatypes(*args):
    
    cint = 0
    cstr = 0
    cbool = 0
    clist = 0
    ctuple = 0
    cdict = 0
    
    ls = []
    for i in args:
        ls.append(i)
        
    for i in range(0,len(ls)):
        if(type(ls[i])==int):
            cint = cint + 1
        elif(type(ls[i])==str):
            cstr = cstr +1
        elif((type(ls[i])==bool)):
            cbool = cbool +1 
        elif((type(ls[i])==list)):
            clist = clist + 1
        elif((type(ls[i])==tuple)):
            ctuple = ctuple + 1
        elif(type(ls[i])==dict):
            cdict = cdict + 1

    return [cint,cstr,cbool,clist,ctuple,cdict]
            
```

```python
count_datatypes(1, 45, "Hi", False)
```

```python
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
```

```python
# A Fibonacci string is a precedence of the Fibonacci series. It works with
# any two characters of the English alphabet (as opposed to the numbers 0 and
# 1 in the Fibonacci series) as the initial items and concatenates them together
# as it progresses in a similar fashion as the Fibonacci series.


def fib_str(
    n,alst):
    res = []
    
    def fib_str_nth(n,alst):
        a = alst[0] #n
        b = alst[1] #k


        if(n==1):
            return a #n
        elif(n==2):
            return b #k
        else:
              #  c = fib_str(n-1,alst) + fib_str(n-2,alst)

            b = fib_str_nth(n-1,alst)
            a = fib_str_nth(n-2,alst)
            c = b + a

        return(c)

    for i in range(1,n+1):
        res.append(fib_str_nth(i,alst))
    
    return ",".join(res)
```

```python
fib_str(6,['n','k'])
```

```python
fib_str(5,['e','a'])
```

```python
# Given an integer between 0 and 26, make a variable (self.answer). That
# variable would be assigned to a string in this format:

def ones_threes_nines(num):
    
    n9 = num//9
    n3 = 0
    n1 = 0
    
    num2 = num - 9*n9
    
    n3 = num2//3
    n1 = (num2 - 3*n3)//1
    
    return "'nines': {}, 'threes': {}, 'ones':{}".format(n9,n3,n1)
    
```

```python
# Write a non-recursive function that takes an integer n and returns the
# Fibonacci sequence&#39;s value at index n.

def fib(num):
    a = 0
    b = 1
    fib_list = [0,1]
    
    for i in range(2,num+1):
        fib_list.append(fib_list[i-1]+fib_list[i-2])
        
    return fib_list[-1]
```
