```python
# Create a function based on the input and output. Look at the examples, there is a pattern

def secret(string):
    
    msg = string.split(".")
    msg = " ".join(msg)
    
    return "<{} class='{}'></{}>".format(msg[0],msg[2:],msg[0])
```

```python
secret("p.one.two.three")
```

```python
secret("p.four.five") 
```

```python
secret("p.one")
```

```python
# Create a function which counts how many lone 1s appear in a given number. 
# Lone means the number doesn't appear twice or more in a row. 

def count_lone_ones(num):
    
    num_list = list(str(num))

    
    count = 0
    
    if(int(num_list[0])==1 and int(num_list[1])!=1):
        count = count + 1
    
    if(int(num_list[-1])==1 and int(num_list[-2])!=1):
        count = count + 1
        
    for i in range(2,len(num_list)-2):
        if(int(num_list[i])==1):
            if((int(num_list[i-1])!=1) and (int(num_list[i+1])!=1)):
                count = count + 1
    return count
```

```python
count_lone_ones(101)
```

```python
count_lone_ones(1191)
```

```python
count_lone_ones(1111) 
```

```python
count_lone_ones(462)
```

## Q3

```python
# Write a method that accepts two integer parameters rows and cols. 
# The output is a 2d array of numbers displayed in column-major order, 
# meaning the numbers shown increase sequentially down each column and wrap to the top of the next column 
# to the right once the bottom of the current column is reached.

def printGrid(rows,cols):
    
    l = [i for i in range(1,rows*cols + 1) ]
    
    res = [[] for j in range(0,rows) ]
    for i in range(0,rows):
    
        for j in range(i,len(l),rows):
            res[i].append(l[j])

    
    for i in range(0,len(res)):
        print(res[i])
```

```python
printGrid(3,6)
```

```python
printGrid(5, 3) 
```

```python
printGrid(4,1)
```

```python
# Given a list of integers, return the smallest positive integer not present in the list.

def min_miss_pos(l):
    
    l.sort(reverse = True)
    l = list(set(l))
    mini = min(l)
    maxi = max(l)

    all_nums = [i for i in range(mini,maxi+1) if i>0]

    
    for i in range(0,len(all_nums)):
        if(all_nums[i] not in l):
            print(all_nums[i])
            break
```

```python
min_miss_pos([5, 9, -2, 0, 1, 3, 9, 3, 8, 9])
```

```python
# Create a function that takes a dictionary of customers, a minimum number of orders and a minimum order price. 
# Return a list of customers that are eligible for a free pizza.

customers = {
  "Batman": [22, 30, 11, 17, 15, 52, 27, 12],
  "Spider-Man": [5, 17, 30, 33, 40, 22, 26, 10, 11, 45]
}

def pizza_points(customers,mino,maxo):
    
    bat_data = customers["Batman"]
    spi_data = customers["Spider-Man"]
    
    countb = 0
    
    for i in range(0,len(bat_data)):
        if(bat_data[i]>=maxo):
            countb = countb + 1

    counts = 0
    
    for i in range(0,len(spi_data)):
        if(spi_data[i]>=maxo):
            counts = counts + 1

    
    if((countb>=mino) and (counts>=mino)):
        return ["Batman","Spider-Man"]
    elif(countb>=mino):
        return ["Batman"]
    elif(counts>=mino):
        return ["Spider-Man"]
    else:
        return []
```

```python
pizza_points(customers, 5, 20) 
```
