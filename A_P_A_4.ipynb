## Q1

```python
# Create a  function fastFib(num) that returns the fibonacci number Fn, of the given num
# as an argument.

def fastFib(num):
    a = 0
    b = 1
    
    if(num==0):
        fn =  a
    elif(num==1):
        fn =  b

    else:
        fn = fastFib(num-1) + fastFib(num-2)
    
    return fn
```

```python
fastFib(10)
```

```python
fastFib(20)
```

## Q2

```python
# Create a function that takes a strings characters as ASCII and returns each
# characters hexadecimal value as a string.

def convert_to_hex(string):
 
    res = ""

    for i in range(0,len(string)):
 
        ch = string[i]

        a = ord(ch)
   
        b = hex(a).lstrip("0x").rstrip("L")

        res += b + " "

    return res
```

```python
convert_to_hex("hello world")
```

## Q3

```python
# Given a censored string and a string of the censored vowels, return the
# original uncensored string.

def uncensor(string, vowels):
    
    res = ""
    count = 0
    for i in range(len(string)):
        if string[i] == "*":
            res = res + vowels[count]
            count = count + 1
        else:
            res = res + string[i]
    
    return res
 
    
```

```python
uncensor("Wh*r* d*d my v*w*ls g*?", "eeioeo")
```

```python
uncensor("abcd","")
```

```python
# Create a function that takes an integer n and returns the factorial of
# factorials. See below examples for a better understanding:


    
def fact_of_fact(n):
  
    def fact(num):
    
        if(num == 0 or num == 1): 
            return 1
    
        else:
            return(num*fact(num-1))
    
    p = 1
    for i in range(0,n+1):
        p = p*fact(i)
    
    return p
```

```python
fact_of_fact(4)
```

```python
fact_of_fact(5)
```
