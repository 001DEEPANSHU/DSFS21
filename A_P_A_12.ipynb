## Q1

```python
def meme_sum(num1, num2):
    
    res = ""
    
    str1 = str(num1)
    str2 = str(num2)
    
    if(len(str1)>len(str2)):
        long = str1
        short = str2
        
    else:
        long = str2
        short = str1
        
    for i in range(0,len(long) - len(short)):
        short = "0" + short
    
    for i in range(0,len(long)):
        res = res + str(int(short[i]) + int(long[i]))

    return int(res)
        
    
```

```python
meme_sum(122,81)
```

```python
meme_sum(1222,30277)
```

## Q2

```python
# Given an integer, create a function that returns the next prime. If the
# number is prime, return the number itself.


def next_prime(num):
    
    def prime(n):
        
        flag = True
        
        if(n==2):
            flag = True
        
        else:
            for i in range(2,n):
                if(n%i==0):
                    flag = False
                    
        return flag
    
    if(prime(num)==True):
        return num
    
    else:
        
        flag = False
        
        while(flag == False):
            num = num + 1
            if(prime(num)==True):
                flag = True
        return num
                
            
            
```

```python
next_prime(24)
```

```python
next_prime(12)
```

```python
next_prime(11)
```

```python
# The Kempner Function, applied to a composite number, permits to find the
# smallest integer greater than zero whose factorial is exactly divided by the
# number.

def kempner(num):
    
    def fact(n):
        if(n==0 or n==1):
            return n
        
        else:
            return n*fact(n-1)

    for i in range(1,fact(num)+1):
        if(fact(i)%num==0):
            return i
        
        
        
```

```python
kempner(6)
```

```python
kempner(10)
```

```python
kempner(2)
```
