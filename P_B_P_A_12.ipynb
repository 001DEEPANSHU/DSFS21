## Q1

```python
# Program to Extract Unique dictionary values

n=int(input("\nEnter no. of data values you want to enter in the dictionary: "))

k=[]

for i in range(1,n+1):
    k.append(input("Enter key {} ".format(i)))

v=[]

for i in range(1,n+1):
    v.append(input("\nEnter value {} ".format(i)))

print("\n")

dic={a:b for a,b in zip(k,v)}
print("Entered dictionarie is: ",dic)

print("\n The unique values in the dictionary are: ",set(v))
```

## Q2

```python
# Program to find the sum of all items in a dictionary values

n=int(input("\nEnter no. of data values you want to enter in the dictionary: "))

k=[]

for i in range(1,n+1):
    k.append(input("Enter key {} ".format(i)))

v=[]

for i in range(1,n+1):
    v.append(int(input("\nEnter value {} ".format(i))))

print("\n")

dic={a:b for a,b in zip(k,v)}
print("Entered dictionarie is: ",dic)

sum = 0 
for i in range(0,n):
    sum = sum + v[i]
    i = i + 1 
print("\nThe sum of all items in the dictionary is: ", sum)
```

## Q3

```python
# Program to merge two dictionaries

n1=int(input("\nEnter no. of data values you want to enter in the first dictionary: "))
n2=int(input("\nEnter no. of data values you want to enter in the second dictionary: "))

k1=[]

for i in range(1,n1+1):
    k1.append(input("Enter key {} ".format(i)))

v1=[]

for i in range(1,n1+1):
    v1.append((input("\nEnter value {} ".format(i))))

print("\n")

dic1={a:b for a,b in zip(k1,v1)}
print("The first dictionary is: ",dic1)


k2=[]

for i in range(1,n2+1):
    k2.append(input("Enter key {} ".format(i)))

v2=[]

for i in range(1,n2+1):
    v2.append((input("\nEnter value {} ".format(i))))

print("\n")

dic2={a:b for a,b in zip(k2,v2)}
print("The second dictionary is: ",dic2)


km= k1+k2
vm = v1+v2

dic_merged={a:b for a,b in zip(km,vm)}
print("The second dictionary is: ",dic_merged)
```

```python
# Program to convert key-values list to flat dictionary

ID = list(range(1,6))
name= ["A"+str(i) for i in range(1,6)]

dic = {"ID" : ID,"Name":name}
print("The dictionary is:\n",dic)
flat_dic = dict(zip(ID,name))

print("\nThe flattened dictonary is:\n",flat_dic)
```

## Q5

```python
# Program to insert an element at the beginning in OrderedDict

from collections import OrderedDict
n=int(input("\nEnter no. of data values you want to enter in the ordered dictionary: "))

k=[]

for i in range(1,n+1):
    k.append(input("Enter key {} ".format(i)))

v=[]

for i in range(1,n+1):
    v.append((input("\nEnter value {} ".format(i))))

print("\n")

dic={a:b for a,b in zip(k,v)}
ordered_dic = OrderedDict(dic)
print("Entered ordered dictionary is: ",ordered_dic)
kn = input("\nEnter key for the new entry")
vn = input("\nEnter value for the new entry")
ordered_dic.update({kn:vn})
print("\nUpdated dictionary is: ",ordered_dic)
```

## Q6

```python
#Program to check order of character in string using OrderedDict()

from collections import OrderedDict
string = str(input("Enter the string: "))
char = str(input("Enter the character: "))

def order_func(string,char):
    ordered_dict = OrderedDict.fromkeys(string)

    i = 0 
    for k,v in ordered_dict.items():
        if(k==char[i]):
            i = i +1

        if(i==len(char)):
            return 'Order of the character is correct'
    
    return'Order of the character is not correct'

order_func(string,char)
```

## Q7

```python
# Program to sort a dictionary by keys

n=int(input("\nEnter no. of data values you want to enter in the dictionary: "))

k=[]

for i in range(1,n+1):
    k.append(input("Enter key {} ".format(i)))

v=[]

for i in range(1,n+1):
    v.append(input("\nEnter value {} ".format(i)))

print("\n")

dic={a:b for a,b in zip(k,v)}
print("Entered dictionary is: ",dic)

print("\nSorted dictionary is: ")

for key in sorted(dic.keys()):
    print(key,":", dic[key])
```
