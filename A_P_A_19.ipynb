```python
# Create a checker board generator, which takes as inputs n and 2 elements to generate an n x n checkerboard
# with those two elements as alternating squares.

def checker_board(n,a,b):
    
    import math
    
    if(a==b):
        return "invalid"
    
    else:
        
        k = math.ceil((n**2)/2)
        
        temp_arr = [a,b]*k
        
        res = [ [] for i in range(0,n)]
        
        for i in range(0,n):
            res[i] = temp_arr[i:i+n]        
    
        return res
    
```

```python
checker_board(3,"A","B")
```

```python
checker_board(4,"c","d")
```

```python
checker_board(4,"c","c")
```

```python
# Create a function that returns True if a string is an almost-palindrome and False otherwise.

def almost_palindrome(string):
    
    import math

    l = len(string)
    
    k = math.ceil(l/2)
    
    if(l%2!=0):
        A= string[0:k]
        B= string[k-1:][::-1]
    else:
        A= string[0:k]
        B= string[k:][::-1]
    
    count = 0
    
    for i in range(0,len(A)):
        if(A[i]!=B[i]):
            count = count + 1
            if(count>1):
                return False
                break

            
    return True
```

```python
almost_palindrome("abcdcbg")
```

```python
almost_palindrome("abcdaaa") 
```

```python
almost_palindrome("1234312") 
```

```python
# Create a function that finds how many prime numbers there are, up to the given integer.

def prime_numbers(num):
    
    res = [i for i in range(2, num+1) if all(i%j!= 0 for j in range(2, i))]
    
    return len(res)
```

```python
prime_numbers(10)
```

```python
prime_numbers(20)
```

```python
prime_numbers(30)
```

```python
# Create a function that determines if the list of users is mapped to a list of anonymous names correctly.

def is_correct_aliases(l1,l2):
    
    flag = True
    for i in range(0,len(l1)):
        if((l1[i][0] != l2[i].split(" ")[0][0]) or (l1[i][0] != l2[i].split(" ")[1][0])):
            flag = False
            break
    return flag
    
    
```

```python
is_correct_aliases(["Adrian M.", "Harriet S.", "Mandy T."], ["Amazing Artichoke", "Hopeful Hedgehog", "Marvelous Mouse"])
```
